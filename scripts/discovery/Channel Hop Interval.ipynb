```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
import matplotlib
pgf_with_rc_fonts = {
    "text.usetex": True,
    "text.latex.unicode": True,
    "font.family": "serif",
    "font.serif": [],                   # use latex default serif font
    "font.sans-serif": ["DejaVu Sans"], # use a specific sans-serif font
    "pgf.preamble": [
        r'\usepackage{amsmath}',
        r'\usepackage{gensymb}',
        r'\usepackage{siunitx}',
        r'\DeclareSIUnit{\revolution}{rev}',
    ]
}
matplotlib.rcParams.update(pgf_with_rc_fonts)

import matplotlib.pyplot as plt

import pandas as pd
import os

# Background Color
bg = '#d3e9f8'
```

```python
dataframes = []    
directory = '../../valid/discovery/discovery2/discovery-hop'

# Walk through each subdirectory of working directory
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith("-results.csv"):
            dataframes.append(pd.read_csv(os.path.join(root,file),sep=','))

df = pd.concat(dataframes, axis=0)

data =(df.groupby(['hop-rate', 'pass'])['timestamp'].count()/df.groupby(['hop-rate', 'pass'])['duration'].max())

fig = plt.figure()
ax = plt.subplot(111)

labels = [109, 119, 129, 139, 149, 159, 169, 179, 189, 199]

bp = ax.boxplot(data.unstack(), 1, labels=labels, patch_artist=True)

for box in bp['boxes']:
    box.set_facecolor('None')

opt = bp['boxes'][7]
opt.set_hatch(r'\\')
opt.set_facecolor(bg)
label = f"Optimal CHI (median: {round(data[0.1833].median(),2)} "
opt.set_label(label + r"$\frac{b}{s}$)")
plt.legend(handles=[opt], loc='lower right')

plt.xlabel(r'Channel Hop Interval (CHI) ($TU$)')
plt.ylabel(r'Beacons Per Second (BPS) ($\frac{b}{s}$)')
fig.savefig('chi.pgf')
```
